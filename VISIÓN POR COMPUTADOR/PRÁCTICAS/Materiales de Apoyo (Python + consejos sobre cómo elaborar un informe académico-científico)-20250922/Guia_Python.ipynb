Del mismo modo, los bucles `for` permiten generar listas de elementos de una forma rápida y sencilla. Por ejemplo, a continuación, en una única línea de código inicializamos una lista multiplicando por $2$ los números enteros del $0$ al $9$:


*   Con la función `range` creamos un rango de valores. Al indicar solamente el número $10$, lo que estamos diciendo es que queremos un rango de valores del $0$ al $10-1$, es decir, el $9$ será el último valor de nuestro rango.
*   El bucle `for` va recorriendo dichos valores $[0, ..., 9]$
*   El `2*x` lo único que hace es multiplicar por $2$ los valores recuperados por el bucle `for`
*   Finalmente, al estar todo entre corchetes `[]`, el resultado queda almacenado en una lista.



````python
new_array = [2*x for x in range(10)]
print(new_array)
````

Dos comandos que pueden ser bastante útiles, y que se emplean para iterar sobre listas son `zip` y `enumerate`. Veamos un par de ejemplos relativos a su uso.

````python
list_a = [*range(1,5)] #* es el argument-unpacking operator, que permite desempaquetar el rango.
list_b = [*range(2,6)]
print(list_a)
print(list_b)
````

````python
list_a*list_b
````

Como vemos, los elementos de una lista no se pueden multiplicar de modo directo haciendo el producto de las listas. Para ello, usamos `zip`, que nos permite iterar sobre varias listas en paralelo.

````python
[a*b for a,b in zip(list_a, list_b)]
````

`enumerate` permite iterar sobre índices y elementos de una lista, tal y como vemos a continuación.

````python
for i, (a,b) in enumerate(zip(list_a, list_b)):
  print(i, a, b)
````

Otro ejemplo:

````python
alist = ['a1', 'a2', 'a3']
blist = ['b1', 'b2', 'b3']
for i, (a,b) in enumerate(zip(alist, blist)):
  print(i, a, b)
````

Como curiosidad técnica, si alguien quiere iterar sobre una secuencia, pero no quiere almacenarla en memoria, puede hacer uso de `yield`. `yield` viene a ser como `return`, pero devolviendo un generador. Esto puede ser de utilidad cuando sabemos que nuestra función devolverá un inmenso conjunto de valores que solo tendremos que leer una vez. Véase, por ejemplo, https://stackoverflow.com/questions/231767/what-does-the-yield-keyword-do o https://stackoverflow.com/questions/47789/generator-expressions-vs-list-comprehensions

````python
def iterarLista():
  mylist = range(5)
  myNewList = []
  for i in mylist:
    myNewList.append(i*i)
  return myNewList

def iterarGenerador():
  mylist = range(5)
  for i in mylist:
    yield i*i

miGenerador = iterarGenerador()
print(type(miGenerador))

miLista = iterarLista()
print(type(miLista))

for i in miGenerador:
  print(i)

for i in miLista:
  print(i)
````

## Objetos mutables e inmutables

Como ya mencionamos antes, hay elementos que son inmutables (como una tupla), y otros que son mutables (como las listas). No obstante, debemos saber que un objeto “inmutable” puede “mutar” si contiene objetos “mutables”. Veamos un ejemplo, en donde una tupla de listas es modificada al alterar las listas que la componen:

````python
first_names = ['Fred', 'George', 'Bill']
last_names = ['Smith', 'Jones', 'Williams']
name_tuple = (first_names, last_names)

print(name_tuple)

first_names.append('Igor')

print(first_names)

print(name_tuple)
````

Comentamos todo esto porque el comportamiento de las funciones varía dependiendo de si operan sobre objetos mutables o inmutables. Y esto determina que se comporten como si se pasasen las variables por valor o por referencia.

### Comportamiento de paso por referencia

````python
def foo(bar):
    bar.append(42)
    print(bar)

answer_list = []
foo(answer_list)
print(answer_list)
````

Si `bar` se refiere a un objeto mutable (como una lista), y `foo` cambia su valor, entonces estos cambios serán visibles fuera de la función. Es decir, esto se asemeja al comportamiento de paso por referencia.

### Comportamiento de paso por valor

````python
def foo(bar):
    bar = 'new value'
    print (bar)
    # >> 'new value'

answer_list = []
foo(answer_list)
print(answer_list)
# >> 'old value'
````

Si `bar` se refiere a un objeto inmutable (como un string), lo máximo que puede hacer `foo` es crear una variable interna `bar` y ligarla a algún otro objeto. Esto se asemeja al comportamiento de paso por valor.

Este **comportamiento dual que tiene Python se llama "Call by Object Reference" o "Call by Assignment"**.

Veamos un último ejemplo, que combina todas estas ideas.

````python
def spam(eggs):
        eggs.append(1)
        eggs = [2, 3]
        print(eggs)

ham = [0]
spam(ham)
print(ham)
````

En el ejemplo anterior:


*   Cuando se llama a `spam`, `eggs` y `ham` apuntan al mismo valor: $[0]$. De hecho, usando `id`, es fácil comprobar que ambos objetos tienen el mismo identificador. Es decir, son el mismo objeto.
*  Cuando hacemos `eggs.append(1)`, $[0]$ se convierte en $[0,1]$
*  Pero cuando hacemos `eggs = [2, 3]`, `eggs` pasa a apuntar a una nueva lista en memoria que contiene $[2,3]$. En cambio, `ham` apunta todavía a la lista $[0,1]$.

# Clases

Python es un lenguaje orientado a objetos en donde, como es lógico, existe la posibilidad de definir nuestras propias clases, métodos y objetos.

````python
class Clase():
    def __init__(self, a):
      # __init__() no es un constructor (__new__()): es llamado inmediatamente después de que el objeto haya sido creado y se usa para inicializarlo
      # No olvidéis el self. self no es una palabra reservada: podríais usar this o myself, por ejemplo, aunque no se recomienda por ser una convención fuertemente aceptada.
        self.a = a

    def llamar(self, b):
        return self.a*b

# Clase2 hereda los métodos de Clase
class Clase2(Clase):
    def __init__(self, a, b=2.0):
        super().__init__(a)
        self.b=b

    # Sobreescribe el método llamar()
    def llamar(self, c):
        return self.a*self.b*c

    def __call__(self, c):
        return self.llamar(c)

c = 3
clase2 = Clase2(a = 1)
d = clase2.llamar(c)
print(d)
````

Más ejemplos con clases en Python:

````python
class Animal(object):
    def __init__(self, species, age): # Constructor `a = Animal(‘bird’, 10)`
        self.species = species # Refer to instance with `self`
        self.age = age # All instance variables are public

    def isPerson(self): # Invoked with `a.isPerson()`
        return self.species == 'Homo Sapiens'

    def ageOneYear(self):
        self.age += 1

class Dog(Animal): # Inherits Animal’s methods
    def ageOneYear(self): # Override for dog years
        self.age += 7

a = Dog(species = 'labrador', age = 6)
print(a.isPerson())
a.ageOneYear()
print(a.age)
````

````python
class Point(object):
    def __init__(self,x = 0,y = 0):
        self.x = x
        self.y = y

    def distance(self):
        """Find distance from origin"""
        return (self.x**2 + self.y**2) ** 0.5

p1 = Point(6,8)

# Si p1 es una instancia de P (i.e. un objeto): p1.meth(arg) y P.meth(p1, arg) son formas equivalentes de llamar a un método
print(p1.distance())

print(Point.distance(p1))

print(type(Point.distance))

print(type(p1.distance))
````

En la celda anterior, podemos observar la diferencia entre una función y un método en Python. Si la función pertenece a una clase, es un método. Dicho método es llamado con un objeto de la clase, y tiene la posibilidad de modificar el estado del objeto. En cambio, cuando está fuera de una clase, y es llamada pasando un determinado argumento con el que operar, es una función.

````python
class A(object):
    @staticmethod # si quitamos esta línea daría error
    def stat_meth():
        print("Aqui no estamos pasando self")

a = A()
a.stat_meth()
````

Atributo que, en cualquier caso, si se crea un método para ello, sí puede ser recuperado/accedido.

````python
class Person:
    def __init__(self, name, age):
        self.__private_attribute = 'This is a private attribute.'
        self.name = name

    def get_private_attribute(self):
        return self.__private_attribute

person = Person('John Doe', 30)
print(person.name)
print(person.get_private_attribute())
print(person.__private_attribute)
````

````python
class Person:
    def __init__(self, name, age):
        self.__private_attribute = 'This is a private attribute.'
        self.name = name

    def get_private_attribute(self):
        return self.__private_attribute

    def set_private_attribute(self):
        self.__private_attribute = 'I have just modified a private attribute.'

person = Person('John Doe', 30)
print(person.name)
print(person.get_private_attribute())
print(person.set_private_attribute())
print(person.get_private_attribute())
````

## Código eficiente

En esta sección intentaremos contextualizar un poco por qué NumPy es mucho más rápido que emplear listas de Python y por qué es bueno evitar, dentro de lo posible, el empleo de bucles for.

1.   La vectorización. NumPy divide una tarea en múltiples subtareas y las procesa en paralelo. Esto permite realizar un único (o pocos) acceso(s) a memoria, en lugar de muchos. Con la vectorización, el código subyacente es paralelizado, de modo que la operación puede ser ejecutada sobre múltiples elementos a la vez, en lugar de haciendo un loop sobre ellos procesando un elemento cada vez. El término vectorización, en este caso, hace referencia al uso de código optimizado y pre-compilado, escrito en un lenguaje de bajo nivel, como pudiera ser C, para realizar operaciones matemáticas simultáneamente sobre una secuencia de datos.
2.	La manera en que se almacenan en memoria los NumPy arrays. NumPy arrays son arrays homogéneos a nivel de tipos y densamente empaquetados (contiguos en memoria; $\textit{locality of reference}$). En cambio, las listas de Python son arrays de punteros a objetos esparcidos por la memoria.





Como consecuencia de todo ello, se recomienda encarecidamente hacer uso de las funciones de NumPy para procesado de arrays, y evitar en lo posible el uso de listas y bucles for.

Veamos algunas funciones que pueden resultar de utilidad:

`np.where` es un if-else vectorizado, que permite responder a la pregunta: "dime dónde, en este array, las entradas satisfacen una condición dada". Es, por tanto, una función muy útil para buscar un elemento en un array.

````python
a = np.arange(5,10)
print(a)
np.where(a < 8)
````

El ejemplo anterior nos indica que las posiciones de los valores que son menores que $8$ son la cero (es decir, la primera), la uno y la dos.

Hagamos una prueba para ver la ganancia de velocidad.

````python
a = np.arange(5,10000000)
%timeit np.where(a < 8)
````

````python
def my_where(a, threshold):
  indexes = []
  for i in range(len(a)):
    if a[i] < threshold:
      indexes.append(i)
  return indexes

%timeit my_where(a, 8)
````

En este ejemplo, podemos ver que nuestra versión secuencial usando un bucle for es más de 200 veces más lenta.

En este caso, la versión que no emplea bucles `for` anidados tarda 1000 veces menos que la que sí los emplea.

````python
from dash import Dash, dcc, html, Input, Output, callback

import plotly.express as px

import json
import pandas as pd

external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']

app = Dash(__name__, external_stylesheets=external_stylesheets)

styles = {
    'pre': {
        'border': 'thin lightgrey solid',
        'overflowX': 'scroll'
    }
}

df = pd.DataFrame({
    "x": [1,2,1,2],
    "y": [1,2,3,4],
    "customdata": [1,2,3,4],
    "fruit": ["apple", "apple", "orange", "orange"]
})

fig = px.scatter(df, x="x", y="y", color="fruit", custom_data=["customdata"])

fig.update_layout(clickmode='event+select')

fig.update_traces(marker_size=20)

app.layout = html.Div([
    dcc.Graph(
        id='basic-interactions',
        figure=fig
    ),

    html.Div(className='row', children=[
        html.Div([
            dcc.Markdown("""
                **Hover Data**

                Mouse over values in the graph.
            """),
            html.Pre(id='hover-data', style=styles['pre'])
        ], className='three columns'),

        html.Div([
            dcc.Markdown("""
                **Click Data**

                Click on points in the graph.
            """),
            html.Pre(id='click-data', style=styles['pre']),
        ], className='three columns'),

        html.Div([
            dcc.Markdown("""
                **Selection Data**

                Choose the lasso or rectangle tool in the graph's menu
                bar and then select points in the graph.

                Note that if `layout.clickmode = 'event+select'`, selection data also
                accumulates (or un-accumulates) selected data if you hold down the shift
                button while clicking.
            """),
            html.Pre(id='selected-data', style=styles['pre']),
        ], className='three columns'),

        html.Div([
            dcc.Markdown("""
                **Zoom and Relayout Data**

                Click and drag on the graph to zoom or click on the zoom
                buttons in the graph's menu bar.
                Clicking on legend items will also fire
                this event.
            """),
            html.Pre(id='relayout-data', style=styles['pre']),
        ], className='three columns')
    ])
])


@callback(
    Output('hover-data', 'children'),
    Input('basic-interactions', 'hoverData'))
def display_hover_data(hoverData):
    return json.dumps(hoverData, indent=2)


@callback(
    Output('click-data', 'children'),
    Input('basic-interactions', 'clickData'))
def display_click_data(clickData):
    return json.dumps(clickData, indent=2)


@callback(
    Output('selected-data', 'children'),
    Input('basic-interactions', 'selectedData'))
def display_selected_data(selectedData):
    return json.dumps(selectedData, indent=2)


@callback(
    Output('relayout-data', 'children'),
    Input('basic-interactions', 'relayoutData'))
def display_relayout_data(relayoutData):
    return json.dumps(relayoutData, indent=2)


if __name__ == '__main__':
    app.run(debug=True)
````

````python
from dash import Dash, html, dcc, Input, Output, callback
import pandas as pd
import plotly.express as px

external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']

app = Dash(__name__, external_stylesheets=external_stylesheets)

df = pd.read_csv('https://plotly.github.io/datasets/country_indicators.csv')


app.layout = html.Div([
    html.Div([

        html.Div([
            dcc.Dropdown(
                df['Indicator Name'].unique(),
                'Fertility rate, total (births per woman)',
                id='crossfilter-xaxis-column',
            ),
            dcc.RadioItems(
                ['Linear', 'Log'],
                'Linear',
                id='crossfilter-xaxis-type',
                labelStyle={'display': 'inline-block', 'marginTop': '5px'}
            )
        ],
        style={'width': '49%', 'display': 'inline-block'}),

        html.Div([
            dcc.Dropdown(
                df['Indicator Name'].unique(),
                'Life expectancy at birth, total (years)',
                id='crossfilter-yaxis-column'
            ),
            dcc.RadioItems(
                ['Linear', 'Log'],
                'Linear',
                id='crossfilter-yaxis-type',
                labelStyle={'display': 'inline-block', 'marginTop': '5px'}
            )
        ], style={'width': '49%', 'float': 'right', 'display': 'inline-block'})
    ], style={
        'padding': '10px 5px'
    }),

    html.Div([
        dcc.Graph(
            id='crossfilter-indicator-scatter',
            hoverData={'points': [{'customdata': 'Japan'}]}
        )
    ], style={'width': '49%', 'display': 'inline-block', 'padding': '0 20'}),
    html.Div([
        dcc.Graph(id='x-time-series'),
        dcc.Graph(id='y-time-series'),
    ], style={'display': 'inline-block', 'width': '49%'}),

    html.Div(dcc.Slider(
        df['Year'].min(),
        df['Year'].max(),
        step=None,
        id='crossfilter-year--slider',
        value=df['Year'].max(),
        marks={str(year): str(year) for year in df['Year'].unique()}
    ), style={'width': '49%', 'padding': '0px 20px 20px 20px'})
])


@callback(
    Output('crossfilter-indicator-scatter', 'figure'),
    Input('crossfilter-xaxis-column', 'value'),
    Input('crossfilter-yaxis-column', 'value'),
    Input('crossfilter-xaxis-type', 'value'),
    Input('crossfilter-yaxis-type', 'value'),
    Input('crossfilter-year--slider', 'value'))
def update_graph(xaxis_column_name, yaxis_column_name,
                 xaxis_type, yaxis_type,
                 year_value):
    dff = df[df['Year'] == year_value]

    fig = px.scatter(x=dff[dff['Indicator Name'] == xaxis_column_name]['Value'],
            y=dff[dff['Indicator Name'] == yaxis_column_name]['Value'],
            hover_name=dff[dff['Indicator Name'] == yaxis_column_name]['Country Name']
            )

    fig.update_traces(customdata=dff[dff['Indicator Name'] == yaxis_column_name]['Country Name'])

    fig.update_xaxes(title=xaxis_column_name, type='linear' if xaxis_type == 'Linear' else 'log')

    fig.update_yaxes(title=yaxis_column_name, type='linear' if yaxis_type == 'Linear' else 'log')

    fig.update_layout(margin={'l': 40, 'b': 40, 't': 10, 'r': 0}, hovermode='closest')

    return fig


def create_time_series(dff, axis_type, title):

    fig = px.scatter(dff, x='Year', y='Value')

    fig.update_traces(mode='lines+markers')

    fig.update_xaxes(showgrid=False)

    fig.update_yaxes(type='linear' if axis_type == 'Linear' else 'log')

    fig.add_annotation(x=0, y=0.85, xanchor='left', yanchor='bottom',
                       xref='paper', yref='paper', showarrow=False, align='left',
                       text=title)

    fig.update_layout(height=225, margin={'l': 20, 'b': 30, 'r': 10, 't': 10})

    return fig


@callback(
    Output('x-time-series', 'figure'),
    Input('crossfilter-indicator-scatter', 'hoverData'),
    Input('crossfilter-xaxis-column', 'value'),
    Input('crossfilter-xaxis-type', 'value'))
def update_x_timeseries(hoverData, xaxis_column_name, axis_type):
    country_name = hoverData['points'][0]['customdata']
    dff = df[df['Country Name'] == country_name]
    dff = dff[dff['Indicator Name'] == xaxis_column_name]
    title = '<b>{}</b><br>{}'.format(country_name, xaxis_column_name)
    return create_time_series(dff, axis_type, title)


@callback(
    Output('y-time-series', 'figure'),
    Input('crossfilter-indicator-scatter', 'hoverData'),
    Input('crossfilter-yaxis-column', 'value'),
    Input('crossfilter-yaxis-type', 'value'))
def update_y_timeseries(hoverData, yaxis_column_name, axis_type):
    dff = df[df['Country Name'] == hoverData['points'][0]['customdata']]
    dff = dff[dff['Indicator Name'] == yaxis_column_name]
    return create_time_series(dff, axis_type, yaxis_column_name)


if __name__ == '__main__':
    app.run(debug=True)
````

````python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact


def interactive_plot(amplitude, frequency):
	x = np.linspace(0, 2 * np.pi, 1000)
	y = amplitude * np.sin(frequency * x)
	plt.figure(figsize=(8, 4))
	plt.plot(x, y)
	plt.xlabel('X-axis')
	plt.ylabel('Y-axis')
	plt.title('Interactive Sine Wave')
	plt.grid(True)
	plt.show()


interact(interactive_plot, amplitude=(1, 5, 0.1), frequency=(1, 10, 0.1))
````

````python
import seaborn as sns
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import numpy as np

lrs = np.linspace(0.0001, 1, 10)
loss = np.array([[i + 1] * 10 for i in range(10)])

palete = sns.color_palette('hsv', n_colors=len(lrs)).as_hex()
palete=list(palete)

fig = make_subplots(rows=1, cols=1,shared_xaxes=True,
                    vertical_spacing=0.02,
                    subplot_titles=['Example of dynamic/interactive figure using Seaborn and Plotly']
                    )

for i,_ in enumerate(lrs):

    fig.add_trace(
        go.Scatter(
          x=list(range(1,len(loss[i])+1)), y=loss[i],
          mode='lines+markers',
          name=f"lr={lrs[i]}",
          marker=dict(size=10,color=palete[i]),
          showlegend=True,
                    legendgroup=f"lr={lrs[i]}"

        ),
        row=1, col=1
    )

fig.update_layout(
    title_font=dict(size=24),
    width=1200,
    height=600,
    coloraxis=dict(autocolorscale=True)
)

fig.update_yaxes(showgrid=True,title_text = 'Loss', title_font=dict(size=20), tickfont=dict(size=18))
fig.update_xaxes(title_text='Epochs', title_font=dict(size=20), tickfont=dict(size=18))

fig.show()
````

## Aprendizaje y predicción a partir de datos externos (ejemplo de regresión)

Como parece lógico, scikit-learn no permite solamente operar con los sencillos datasets que ofrece por defecto. Permite cargar y trabajar con conjuntos de datos externos: https://scikit-learn.org/stable/datasets/loading_other_datasets.html#external-datasets

A continuación, se muestra un ejemplo (extraído de los materiales de [Aurélien Géron](https://github.com/ageron/handson-ml3/blob/main/01_the_machine_learning_landscape.ipynb)), de cómo cargar un dataset externo (en este caso en `.csv`) y cómo entrenar y predecir con un modelo de regresión lineal (del mismo modo que en el apartado anterior mostrábamos un problema de clasificación, aquí se muestra un problema de regresión).

````python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Download and prepare the data
data_root = "https://github.com/ageron/data/raw/main/"
lifesat = pd.read_csv(data_root + "lifesat/lifesat.csv")
X = lifesat[["GDP per capita (USD)"]].values
y = lifesat[["Life satisfaction"]].values

# Visualize the data
lifesat.plot(kind='scatter', grid=True,
             x="GDP per capita (USD)", y="Life satisfaction")
plt.axis([23_500, 62_500, 4, 9])
plt.show()

# Select a linear model
regressor = LinearRegression()

# Train the model
regressor.fit(X, y)

# Make a prediction for Cyprus
X_new = [[37655.2]]  # Cyprus' GDP per capita in 2020
print(regressor.predict(X_new)) # outputs [[6.30165767]]
````

Reemplazar regresión lineal por otro modelo de aprendizaje automático, en este caso k-Nearest Neighbors (con $k=3$), resulta tremendamente sencillo.

````python
# Select a 3-Nearest Neighbors regression model
from sklearn.neighbors import KNeighborsRegressor

regressor = KNeighborsRegressor(n_neighbors=3)

# Train the model
regressor.fit(X, y)

# Make a prediction for Cyprus
print(regressor.predict(X_new)) # outputs [[6.33333333]]
````

### Partición simple en entrenamiento y test ($holdout$)

Antes incluso de emplear validación cruzada, vamos a comenzar mostrando cómo se puede realizar una única partición ($holdout$) que separa simplemente un conjunto de entrenamiento y uno de test. Lo vemos con un ejemplo empleando $\textit{random forest}$.

````python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.ensemble import RandomForestClassifier

X, y = datasets.load_iris(return_X_y=True)
X.shape, y.shape
````

En scikit-learn, se puede calcular rápidamente una división aleatoria en conjuntos de entrenamiento y test con la función  `train_test_split`. Lo que vamos a hacer a continuación es separar un 40% de los datos para testear/evaluar nuestro clasificador.

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
X_train.shape, y_train.shape
````

````python
X_test.shape, y_test.shape
````

El 40% de 150 son 60 ejemplos, y vemos que es eso lo que, automáticamente, hemos separado para test.

Ahora empleamos nuestro clasificador:

````python
classifier = RandomForestClassifier(n_estimators=250, max_depth=3, random_state=0)
classifier.fit(X_train, y_train)
````

Y evaluamos el rendimiento en el conjunto de test. Visualizamos la $accuracy$ media sobre dichos datos.

````python
classifier.score(X_test, y_test) #Return the mean accuracy on the given test data and labels.
````

En caso de que quisiésemos extraer más métricas para analizar el rendimiento alcanzado, podríamos emplear el comando `classification_report` (https://scikit-learn.org/stable/modules/generated/sklearn.metrics.classification_report.html).

````python
from sklearn.metrics import classification_report
pred = classifier.predict(X_test)
print(classification_report(y_test, pred, target_names = iris.target_names))
````

También se puede calcular de forma inmediata la matriz de confusión (https://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html y https://scikit-learn.org/stable/modules/generated/sklearn.metrics.ConfusionMatrixDisplay.html)

````python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test, pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=iris.target_names)
disp.plot()
````

### Validación cruzada ($\textit{cross-validation}$)

A continuación, vemos cómo se podría estimar el rendimiento de nuestro clasificador empleando $\textit{5-fold cross-validation}$. Para ello, hacemos uso de la función `cross_val_score` (https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_score.html#sklearn.model_selection.cross_val_score). Esta función representa la forma más sencilla de emplear $\textit{cross-validation}$ para calcular el error de un estimador. Dicha función lo que necesita es el `estimator` y los datos.

````python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(classifier, X_train, y_train, cv=5)
scores
````

````python
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
````

Resulta fundamental revisar la documentación para comprender el comportamiento de estas funciones. Por ejemplo, ¿qué son esos `scores`? Si consultamos la documentación relativa a `cross_val_score` veremos que son un `Array of scores of the estimator for each run of the cross validation.` Es decir, se trata del valor de las métricas calculadas sobre los conjuntos de test de cada $fold$/$split$ de $\textit{cross-validation}$. Pero, ¿qué metrica concreta se está aplicando en cada caso? Aquí estamos ante un problema de clasificación, ¿cómo sabemos que es la $accuracy$ la que se muestra como `score`? De nuevo, es necesario revisar la documentación para saberlo. Esta indica que `scoring: str or callable, default=None [...] If None, the estimator’s default scorer (if available) is used.` En este caso, estamos empleando $\textit{Random Forest}$, y la ayuda indica que el método `score()` devuelve `the mean accuracy on the given test data and labels.`

Si estuviésemos ante un problema de regresión, y empleásemos regresión lineal (https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html) para resolverlo, veríamos que el `score` por defecto sería el coeficiente de determinación ($R^2$) (`Return the coefficient of determination of the prediction`). Como decíamos, es importante revisar la documentación para poder comprender los resultados que obtenemos. Si seguimos con el ejemplo de regresión, veremos que, en caso de que queramos evaluar el rendimiento empleando el error cuadrático medio ($\textit{mean squared error}$, MSE), la alternativa que nos ofrece el parámetro `scoring` dentro de `cross_val_score` (https://scikit-learn.org/stable/modules/model_evaluation.html#the-scoring-parameter-defining-model-evaluation-rules) es `neg_mean_squared_error`. Por lo que tendremos que negar ese valor para obtener el MSE.  

Es importante hacer notar que `cross_val_score` solo admite una única métrica en `scoring`. En caso de incluir más de una (lo cual, en la mayor parte de circunstancias es adecuado, dado que no es lo más recomendable evaluar el rendimiento de un modelo usando una única métrica) obtendríamos un error: `ValueError: For evaluating multiple scores, use sklearn.model_selection.cross_validate instead.` Otra lección importante de esto es la siguiente: prestad atención a los errores que os proporciona scikit-learn, dado que estos os pueden dar pistas claras sobre cómo solucionarlos.

Existen numerosas funciones relacionadas y que pueden ser de gran utilidad en distintos contextos:

* `sklearn.model_selection.cross_validate` (https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_validate.html): permite realizar lo mismo que `cross_val_score` pero, entre otras cosas, empleando más de una métrica en el parámetro `scoring` (véase https://scikit-learn.org/stable/modules/cross_validation.html#the-cross-validate-function-and-multiple-metric-evaluation).

*   `sklearn.model_selection.KFold` (https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.KFold.html): permite obtener los índices para los distintos conjuntos de entrenamiento y test. Es decir, permite tener acceso directo a los folds y operar con ellos como mejor se considere.
*   `sklearn.model_selection.StratifiedKFold` (https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.StratifiedKFold.html): realiza lo mismo que la función anterior, pero proporcionando particiones estratificadas. Es decir, los folds generados conservan el porcentaje de muestras en cada clase.
* `sklearn.model_selection.ShuffleSplit` (https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.ShuffleSplit.html): realiza $cross-validation$ de forma aleatoria, sin asegurar/forzar que los distintos folds sean disjuntos.
* también existen códigos de ejemplo para visualizar, y comprender mejor, el comportamiento de los distintos algoritmos de validación cruzada en scikit-learn: https://scikit-learn.org/stable/auto_examples/model_selection/plot_cv_indices.html

Veamos cómo se podría usar, por ejemplo, `ShuffleSplit`:



````python
from sklearn.model_selection import ShuffleSplit
n_samples = X_train.shape[0]
cv = ShuffleSplit(n_splits=5, test_size=0.3, random_state=0)
cross_val_score(classifier, X_train, y_train, cv=cv)
````

## Preprocesado de datos, Pipelines y otras utilidades

scikit-learn incluye también herramientas para transformar los datos de entrada: https://scikit-learn.org/stable/modules/preprocessing.html. A continuación, mostramos un ejemplo en donde escalamos (en este caso, $standardize$) datos restando la media y dividiendo por la desviación típica. De este modo, nuestros datos tendrán media cero y desviación típica 1.

````python
from sklearn import preprocessing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)

# Escalamos los datos de entrenamiento
scaler = preprocessing.StandardScaler().fit(X_train)
X_train_transformed = scaler.transform(X_train)

# Entrenamos un modelo de aprendizaje automático sobre estos datos
clf = svm.SVC(C=1).fit(X_train_transformed, y_train)

# Ahora hacemos inferencia sobre datos de test. Para ello, debemos escalar los
# datos de test siguiendo exactamente el mismo proceso realizado en entrenamiento
X_test_transformed = scaler.transform(X_test)
clf.score(X_test_transformed, y_test)
````

El rendimiento es muy inferior, dado que nuestro modelo ha sido entrenado con los datos transformados de un modo, y ahora le presentamos datos de test en otro "formato".

Como es lógico, podemos aplicar este preprocesado a la matriz que queramos. Tal y como vemos en la siguiente celda, aplicamos la normalización a una matriz creada por nosotros de tamaño 5$\times$3, e imprimimos la media y desviación típica por columna antes y después de dicho proceso de escalado:

````python
from sklearn.preprocessing import StandardScaler
import numpy as np

data = np.array([[ 1., -1.,  2.],
	            [ 2.,  0.,  0.],
	            [ 0.,  1., -1.],
              [ 3.,  3.,  3.],
              [ 4.,  0.,  0.]])

print('Mean and std before normalization')
print(data.mean(axis=0))
print(data.std(axis=0))

scaler = StandardScaler()
scaler.fit(data)

data_new = scaler.transform(data)

print('Mean and std after normalization')
print(data_new.mean(axis=0))
print(data_new.std(axis=0))
````

Por otro lado, un Pipeline es una herramienta que ofrece scikit-learn para facilitar la composición de estimadores. Por ejemplo, las dos líneas de código siguientes (si ignoramos el `import`) permiten encadenar/componer preprocesado y entrenamiento y, a continuación, evaluar empleando $\textit{5-fold cross-validation}$ como protocolo de validación experimental.

````python
from sklearn.pipeline import make_pipeline
clf = make_pipeline(preprocessing.StandardScaler(), svm.SVC(C=1))
cross_val_score(clf, X, y, cv=cv)
````

Los Pipelines, tal y como indica la documentación (https://scikit-learn.org/stable/modules/compose.html), resultan de gran utilidad, dado que permiten:
* encadenar/componer operaciones complejas en unas pocas líneas de código,
* incorporar la selección de valores para los hiperparámetros dentro del proceso,
* y evitar errores que se podrían cometer a la hora de implementar todo de forma separada (por ejemplo, $\textit{data snooping}$).

En el ejemplo que vemos a continuación, lo que hacemos es encapsular, por medio de un Pipeline, la normalización de datos y la estimación de hiperparámetros empleando $\textit{grid search}$ (https://scikit-learn.org/stable/modules/grid_search.html). En concreto, empleamos regresión logística (https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html) y buscamos estimar un buen valor para el parámetro $C$ (factor de regularización; a menor valor, más fuerte la regularización) empleando $\textit{3-fold cross-validation}$. Luego, el mejor modelo encontrado es evaluado también por medio de validación cruzada.

````python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

clf = make_pipeline(StandardScaler(),
                    GridSearchCV(LogisticRegression(),
                                 param_grid={'C': [0.001, 0.01, 0.1, 1]},
                                 cv=3))

cross_val_score(clf, X, y, cv=cv)
````

Como es lógico, esta estimación de hiperparámetros no tiene que emplearse, forzosamente, a través de un Pipeline. A continuación, se muestra un ejemplo en donde se aplica un $\textit{grid search}$ (que emplea $\textit{5-fold cross-validation}$ por defecto) sobre los datos de entrenamiento para escoger una buena configuración de nuestro SVM (parámetro de regularización y kernel a emplear). Imprimimos los errores medios (obtenidos en test) de cada una de las 20 configuraciones probadas, así como las configuraciones en sí (para poder asociar errores y configuraciones). Finalmente, mostramos los mejores parámetros encontrados, realizamos la predicción sobre los datos de test, y mostramos el $classification\_report$ asociado.

````python
from sklearn	import	svm,	datasets
from sklearn.model_selection	import	GridSearchCV,	train_test_split
from sklearn.metrics	import	classification_report

iris	=	datasets.load_iris()
parameters	=	{'kernel':('linear', 'poly', 'rbf', 'sigmoid'),	'C':[0.001, 0.01, 0.1, 1,	10]}
svc	=	svm.SVC()
X	=	iris.data
y	=	iris.target
X_train,	X_test,	y_train,	y_test	=	train_test_split(X,	y, test_size=0.25,	random_state=0)

clf	=	GridSearchCV(svc,	parameters)
clf.fit(X_train,y_train)

print(clf.cv_results_['mean_test_score'])
print(clf.cv_results_['params'])
print('BEST PARAMS FOUND: ', clf.best_params_)


y_pred	=	clf.predict(X_test)
print(classification_report(y_test,	y_pred))
````
