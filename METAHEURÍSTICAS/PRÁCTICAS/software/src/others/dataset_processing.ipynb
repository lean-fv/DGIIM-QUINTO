```python
import pandas as pd
```

```python
DIR='../../datos_portfolio_2526'
CSVs=['ibex_35', 'syp_100', 'syp_500']
```

```python
for csv in CSVs:
    name=f'{DIR}/{csv}'
    df = pd.read_csv(name+'.csv')
    # 2. Ensure the 'Date' column is interpreted as a real date (datetime)
    # This is vital to easily extract the year
    df['Date'] = pd.to_datetime(df['Date'])

    # 3. Filter to keep only the rows where the year is 2025
    df_2025 = df[df['Date'].dt.year == 2025]

    # Optional: If you also want a Train dataset (2015-2024), it would be like this:
    df = df[df['Date'].dt.year < 2025]

    # 4. Save the result to a new CSV file
    # index=False prevents saving the row numbers column (0, 1, 2...) on the left
    df_2025.to_csv(name+'_2025.csv', index=False)
    df.to_csv(name+'.csv', index=False)

    print(f"Original dataset: {len(df)} rows.")
    print(f"2025 dataset (Test): {len(df_2025)} rows extracted and saved.")
```
